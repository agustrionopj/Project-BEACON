```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
import pandas as pd
import pymc3 as pm
import arviz as az
import corner
import astropy.units as u

import theano
import theano.tensor as tt
from theano.compile.ops import as_op

az.style.use("arviz-darkgrid")
```

```python
# λ_range = np.linspace(5000,9000,100)*u.Angstrom
λ_range = np.arange(5500, 9000, 100)*u.Angstrom
x = np.linspace(-100,100,1000)
A_true = 100.
d = 3.84e5*u.km
λ_range
```

```python
y0 = np.zeros(len(x))

for i in range (len(λ_range)):
    w = x*u.m * np.sqrt(2 / (λ_range[i] * d))
    fre_num = w.decompose() * 1. * u.dimensionless_unscaled
    cc, ss = sc.fresnel(fre_num.value)
    # y = 0.5*((cc + 0.5)**2 + (ss + 0.5)**2)
    y = A_true*0.5*((cc + 0.5)**2 + (ss + 0.5)**2)
    y0 += y
    
# y0 = y0/np.mean(λ_range)
```
