```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
import pandas as pd
import pymc3 as pm
import arviz as az
import corner
import astropy.units as u

import theano
import theano.tensor as tt
from theano.compile.ops import as_op

az.style.use("arviz-darkgrid")
```

```python
# λ_range = np.linspace(5000,9000,100)*u.Angstrom
λ_range = np.arange(5500, 9000, 100)*u.Angstrom
x = np.linspace(-100,100,1000)
A_true = 100.
d = 3.84e5*u.km
λ_range
```

```python
def poli_func(x, d, A_true):
    y0 = np.zeros(len(x))
    for i in range (len(λ_range)):
        w = x*u.m * np.sqrt(2 / (λ_range[i] * d))
        fre_num = w.decompose() * 1. * u.dimensionless_unscaled
        cc, ss = sc.fresnel(fre_num.value)
        # y = 0.5*((cc + 0.5)**2 + (ss + 0.5)**2)
        y = A_true*0.5*((cc + 0.5)**2 + (ss + 0.5)**2)
        y0 += y
    return y0
    
# y0 = y0/np.mean(λ_range)
```

```python
y00 = poli_func(x, d, A_true)
```

```python
y00.shape
```

```python
plt.style.use("ggplot")

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(x, y00)
```

```python
import numpy as np
import theano
from theano import tensor as tt
import pymc3 as pm
import scipy
```

```python
def diffract(t, t0, A, v):
    cc, ss = scipy.special.fresnel(v*(t-t0))
    y = A*0.5*((cc + 0.5)**2 + (ss + 0.5)**2)
    return y
```

```python
x_data = np.linspace(-5, 5, 200)
σ_true = 10.
A_true = 100.
t0_true = 0.
v_true = 1.0

y_data = diffract(x_data, t0_true, A_true, v_true)
y_data = np.random.normal(y_data, scale=σ_true)
```

```python
with pm.Model() as model_fresnel:
    amp = pm.Normal('A', mu=90, sigma=20)
    σ = pm.HalfCauchy('σ', beta=1)
    t0 = pm.Normal('t0', mu=0., sigma=20)
    v = pm.Normal('v', mu=1., sigma=10)
    
    mu = diffract(x_data, t0, amp, v) # model-nya (expected value)
    # y_pred = pm.Normal('y_pred', mu=mu, sd=σ, observed=y_data) # likelihood-nya

pm.model_to_graphviz(model_fresnel)
amp.shape
```
